```python
telco_df['Churn'] = telco_df['Churn'].replace({'Yes': 1, 'No': 0})
```

```python
def uniplot(df, col, title, hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30

    temp = pd.Series(data=hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4 * len(temp.unique())
    fig.set_size_inches(width, 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)

    # Ensure the 'gender' column values are consistent
    df['gender'] = df['gender'].str.lower()  # Convert to lowercase for consistency

    # Convert 'gender' to categorical
    df['gender'] = pd.Categorical(df['gender'], categories=['female', 'male'], ordered=True)

    # Specify the order of categories for 'gender'
    order = df[col].value_counts().index if hue else None

    # Use a custom color palette with lowercase keys
    custom_palette = {'female': 'blue', 'male': 'orange'}

    ax = sns.countplot(data=df, x=col, order=order, hue=hue, palette=custom_palette)

    # Adding data labels
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2., height, f'{height}', ha="center", va="bottom")

    # Adding labels
    ax.set_xlabel(col.capitalize())
    ax.set_ylabel('Count (log scale)')

    plt.show()

# Example usage:
# Assuming 'telco_df_Churned' and 'telco_df_Active' are your DataFrames
uniplot(telco_df_Churned, col='Partner', title='Partner vs Gender for Churned Customers', hue='gender')
uniplot(telco_df_Active, col='Partner', title='Partner vs Gender for Active Customers', hue='gender')
```

#Numerical Analysis

```python
telco_df.head(5)
```

```python
telco_df_dummies = pd.get_dummies(telco_df)

telco_df_dummies.head(5)
```

```python
sns.lmplot(data=telco_df_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

```python
Mth = sns.kdeplot(telco_df_dummies.MonthlyCharges[(telco_df_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(telco_df_dummies.MonthlyCharges[(telco_df_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

**Insights** : Churn is high when

1.   Monthly charges are high




```python
Tot = sns.kdeplot(telco_df_dummies.TotalCharges[(telco_df_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = sns.kdeplot(telco_df_dummies.TotalCharges[(telco_df_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

Insights : Churn is high when
When total charges are low
